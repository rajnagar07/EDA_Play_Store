```python
#1mb = 1024 kb 
def size_process(item):
    if str(item[-1]) == "M":
        res = float(str(item).replace("M",""))
        res*=1024  #basically converting the mb into the kb
        return res
    elif str(item[-1]) == "K":
        res = float(str(item).replace("K",""))
        return res
    else:
        return str(np.nan)
```

```python
size_process("19K")
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```
